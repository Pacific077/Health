```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://health.gov/our-work/national-health-initiatives/health-literacy/consumer-health-content/free-web-content/apis-developers/api-content')
```

```python
webpage = driver.page_source
```

```python
driver.quit()
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
tbody_elements = soup.find_all('tbody')
```

```python
tbody_elements
```

```python
tbody_elements[0]
```

```python
tr_elements = tbody_elements[0].find_all('tr')
```

```python
tr_elements
```

```python
td_data_list = []
```

```python
for tr in tr_elements:
    td_elements = tr.find_all('td')
    if len(td_elements) == 2:
        td_number = td_elements[0].text.strip()
        td_sentence = td_elements[1].text.strip()
        td_data_list.append({"id": td_number, "title": td_sentence})
```

```python
for td_data in td_data_list:
    print(td_data)
```

```python
import json
```

```python
with open('td_data.json', 'w') as json_file:
    json.dump(td_data_list, json_file, indent=2)
```
